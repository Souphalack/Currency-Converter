# 1. State the problem Clearly
This is a Python currency converter application. It will convert American dollars or Japanese Yen to Australian dollars. It will also allow the user
to choose a currency, value and display its value in AUD.

* User can choose the type of currency between US Dollar or Japanese Yen.
* User can input one number.
* The program will convert from the original currency according to user input and multiply by the exchange rate to convert it to new currency in Australian Dollar.
* The program will ask if user want to perform another conversion or exist.


# 2. Describe the Input and Output information
## Input
* Screen shows available currency
* Choose option 1 or 2
* Input a number, it can be a whole number with or without a decimal point.
* Yes/No options. If Yes, the program will perform another conversion. If No, the program will exit.

## Output
* Display the result by calculating the converted amount as the output
* Prompt to perform another conversion, have options to continue or exit the program.

# 3. Work the problem by hand
* Exchange rate for USD to AUD is: 1 USD = 1.38 AUD.
* Exchange rate for YEN to AUD is: 1 YEN = 0.010 AUD.

### USD to AUD 
| Example               | Expected |
|------|------|
|USD 0     * 1.38         | AUD 0      | 
|USD 100   * 1.38         | AUD 138    | 
|USD 255.5 * 1.38         | AUD 352.59 |
|USD 1500 * 1.38          | AUD 2070   |

### YEN to AUD 
| Example                  | Expected|
|------|------|
|YEN 0     * 0.010         | AUD 0     | 
|YEN 1000  * 0.010         | AUD 10    | 
|YEN 15750 * 0.010         | AUD 157.5 |
|YEN 52500 * 0.010         | AUD 525   |

This conversion focus on valid inputs, it doesn't consider the potential invalid inputs if user insert a number in String.

# 4. Develop an algorithm and Convert it to Python

* Promt user: Which currency you would like to convert: 1. USD - AUD or 2. YEN - AUD
* Get the answer from user and store into a variable
* Verify what the user typed

* If the user type 1

  Prompt the user the amount of USD they want to convert

  Store what the user typed into a variable 

  AUD = USD amount * 1.38

  Output amount to user
  
  run do again ()


* If the user type 2
    
  Prompt the user the amount of YEN they want to convert

  Store what the user typed into a variable 

  AUD = YEN amount * 1.38

  Output amount to user

  run do again ()

* If the User Typed anything else

  Tell the user what they did wrong

  Run the scriipt again ()

* Ask The user if want to do again

  If choice 1

  If choice 2

```python
# Currency Converter program
!pip -q install install PyInputPlus
import pyinputplus as pyip

def currency_converter():
  ''' This programe will convert American dollars or Japanese Yen to Australian dollars. It allows the user
to choose a currency, value and display its value in AUD.
  '''
  # Prompt user: Which currency you would like to convert: 1. USD - AUD or 2. YEN - AUD
  # Get the answer from user and store into a variable
  user_choice = input('What currency would you like to convert? \n 1. USD to AUD \n 2. YEN to AUD\n')

  # If the user type 1
  if user_choice == "1":
    # Store what the user typed into a variable
    # Prompt the user the amount of USD they want to convert
    user_usd = float(input ("Enter the amount in USD you wish to convert \n"))
  

    # AUD = USD amount * 1.38
    aud = round(user_usd * float(1.38), 2)
    # Output amount to user
    print(f"USD ${user_usd} = AUD ${aud:.2f}") # Stack Overflow, Retrive from: https://stackoverflow.com/questions/455612/limiting-floats-to-two-decimal-points
    print(" ")
    do_again()

    # If the user type 2
  elif user_choice == "2":
    # Store what the user typed into a variable
    # Prompt the user the amount of YEN they want to convert
    user_yen = float(input ("Enter the amount in YEN you wish to convert \n"))

    # AUD = YEN amount * 0.010
    aud = round(user_yen * float(0.010), 2)
    # Output amount to user
    print(f"YEN {user_yen} = AUD ${aud:.2f}")
    print(" ")
    do_again()

  # If the User Typed anything else
  else:
    # Tell the user what they did wrong
    print("Error: You entered the invalid information. Please try again ")
    print("--------------------------------------------------")
  # Run the scriipt again
    currency_converter()

def do_again():
  # Prompt the user if they would like to run the currency converter again
  user_do_again = input("Would you like to convert again?\n y \n n \n")
  print("--------------------------------------------------")

  # Check what the user typed
  # If choice 1
  if user_do_again == "y":
    currency_converter()
  # If choice 2
  elif user_do_again == "n":
    print ("Thank you for using the Currency Conveter")

  # If choice was anything else
  else:
    print ("Error You entered the invalid information. Please try again")
    do_again()
  print("-----------------------------------------------------")


#Call a function
currency_converter() 
```
